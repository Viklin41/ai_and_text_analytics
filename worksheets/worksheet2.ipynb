```python
data.index = data['Region']
num_data = data.loc[:,'Assamese':'Urdu']

data_centered = num_data - num_data.mean()

data_standardized = data_centered / num_data.std()

data_standardized
```

```python
data_standardized.transpose()
```

```python
import numpy as np
corr_matrix = np.cov(data_standardized.T)
pd.DataFrame(corr_matrix)
```

```python
eigvals, eigvecs = np.linalg.eig(corr_matrix)
```

```python
idx = eigvals.argsort()[::-1]  # indices of eigenvalues from largest to smallest
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]
```

```python
PC_scores = data_standardized.values @ eigvecs  # shape: (31 regions × 22 PCs)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(PC_scores[:, 0], PC_scores[:, 1])

# annotate regions
for i, region in enumerate(data_standardized.index):
    plt.text(PC_scores[i, 0], PC_scores[i, 1], region, fontsize=8)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of Indian Languages by Region')
plt.show()
```
